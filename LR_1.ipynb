health – текущее состояние здоровья (от 1 – очень плохое до 5 – очень

* G1 – оценка за первый период (от 0 до 20);
* G2 – оценка за второй период (от 0 до 20);

## 7. Создайте новый датафрейм fam, выполнив срез от столбца school до  столбца famrel и удалив все столбцы от traveltime до romantic включительно (оставив только характеристики семьи)

## 8. Определите процент городских и сельских студентов в каждой школе

```python
address_percentage = df.groupby('school')['address'].value_counts(normalize=True).mul(100).unstack(fill_value=0)
print(address_percentage.round(2))
```

## 9. У какого процента студентов (по школам) и мать и отец имеют высшее  образование?

```python
higher_edu_parents = df[(df['Medu'] == 4) & (df['Fedu'] == 4)]
percentage_higher_edu = higher_edu_parents.groupby('school').size() / df.groupby('school').size() * 100
print(percentage_higher_edu.fillna(0).round(2))
```

## 10. У какого процента студентов (по школам) и мать и отец вообще не  имеют образования?

```python
no_edu_parents = df[(df['Medu'] == 0) & (df['Fedu'] == 0)]
percentage_no_edu = no_edu_parents.groupby('school').size() / df.groupby('school').size() * 100
print(percentage_no_edu.fillna(0).round(2))
```

## 11. Какой процент родителей, имеющих высшее образование, проживает  совместно?

```python
parents_with_any_higher_edu = df[(df['Medu'] == 4) | (df['Fedu'] == 4)]
living_together_among_higher_edu = (parents_with_any_higher_edu['Pstatus'] == 'T').mean() * 100
print(f"{living_together_among_higher_edu:.2f}%")
```

## 12. Определите процент семей, в которых один из родителей имеет высшее  образование, а другой не имеет, но проживают совместно

```python
condition_one_higher_one_none = (
    ((df['Medu'] == 4) & (df['Fedu'] == 0)) |
    ((df['Medu'] == 0) & (df['Fedu'] == 4))
)
condition_living_together = (df['Pstatus'] == 'T')

target_families = df[condition_one_higher_one_none & condition_living_together]
percent_target_families = (len(target_families) / len(df)) * 100
print(f"{percent_target_families:.2f}%")
```

## 13. Какое качество семейных отношений в этих семьях?

```python
print(target_families['famrel'].describe().round(2))
```

## 14. У скольких студентов работа одного или обоих родителей связана с  гражданскими услугами (административными или полицейскими)?

```python
civil_services_parents_students = df[
    (df['Mjob'] == 'services') | (df['Fjob'] == 'services')
]
print(f"Количество студентов: {len(civil_services_parents_students)}")
```
