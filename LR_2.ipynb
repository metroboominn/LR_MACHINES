```python
top_data = data[["Company", "Num_Sites"]]
top_data = top_data.groupby("Company").sum()
top_data = top_data.sort_values("Num_Sites", ascending=False)
top_data = top_data[:2].index.values
sns.boxplot(y="Num_Sites", x="Company",
           data=data[data.Company.isin(top_data)], hue="Company")
```

```python
c = data["Churn"].map({0: "green", 1: "red"})
edge_c = data["Churn"].map({0: "lightgreen", 1: "darkred"})
plt.scatter(
    data["Years"],
    data["Total_Purchase"],
    color=c,
    edgecolors=edge_c
)
plt.xlabel("Количество лет клиента с компанией")
plt.ylabel("Общее количество покупок")
plt.title("Распределение клиентов по длительности сотрудничества и числу покупок")
plt.show()
```

```python
data_no_churn = data[data["Churn"] == 0]
data_churn = data[data["Churn"] == 1]
plt.scatter(
    data_no_churn["Years"],
    data_no_churn["Total_Purchase"],
    color="green",
    edgecolors="lightgreen",
    label="Остались"
)
plt.scatter(
    data_churn["Years"],
    data_churn["Total_Purchase"],
    color="red",
    edgecolors="darkred",
    label="Ушли"
)
plt.xlabel("Количество лет клиента с компанией")
plt.ylabel("Общее количество покупок")
plt.title("Распределение клиентов по длительности сотрудничества и числу покупок")
plt.legend()
plt.show()
```

## 7. Корреляция признаков
В реальных задачах анализа оттока клиентов при первичном анализе данных важно выявить взаимосвязи между количественными признаками. В пакете Pandas для этого имеется встроенный инструмент — метод corr() класса DataFrame, который позволяет получить матрицу корреляций признаков. Ниже приведён пример его использования для набора данных о клиентах:

```python
data.corr(numeric_only=True)
```

Полученная матрица корреляций имеет небольшой размер, соответствующий количеству числовых признаков в наборе данных о клиентах (Age, Total Purchase, Years, Num sites, Account Manager, Churn). В реальных задачах анализа оттока клиентов матрицы корреляции могут быть гораздо больше, и анализировать их вручную крайне трудоемко. Для выделения сильных взаимосвязей лучше использовать визуализацию с помощью heatmap.

```python
sns.heatmap(data.corr(numeric_only=True), cmap=plt.cm.PuBuGn, annot=True)
plt.show()
```

Коррелирующие признаки обычно удаляются или не учитываются при построении моделей прогнозирования оттока клиентов. Из карты heatmap видно, что некоторые признаки взаимосвязаны: например, может наблюдаться умеренная корреляция между Total Purchase и Years, а также слабая отрицательная корреляция между Num sites и Total Purchase. В таких случаях один из сильно коррелирующих признаков можно удалить, чтобы избежать избыточности данных.

```python
high_corr = ["Total_Purchase"]
data_uncorr = data.drop(high_corr, axis=1)
```

```python
sns.heatmap(data_uncorr.corr(numeric_only=True), cmap=plt.cm.rainbow, annot=True)
plt.show()
```
